```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"user_behavior_dataset.csv")
df
```

```python
df.rename(columns={'Device Model': 'Dòng thiết bị', 
                   'Operating System': 'HĐH', 
                   'Gender': 'Giới tính', 
                   'Age': 'Tuổi', 
                   'Battery Drain (mAh/day)': 'Mức độ hao pin (mAh/ngày)', 
                   'Number of Apps Installed': 'Số lượng ứng dụng cài đặt', 
                   'Data Usage (MB/day)': 'Mức độ sử dụng dữ liệu (MB/ngày)', 
                   'App Usage Time (min/day)': 'Thời gian sử dụng ứng dụng (phút/ngày)', 
                   'Screen On Time (hours/day)': 'Thời gian sử dụng màn hình (giờ/ngày)'}, 
                   inplace=True)

df['Giới tính'] = df['Giới tính'].replace({'Male': 'Nam', 'Female': 'Nữ'})

dev_count = df['Dòng thiết bị'].value_counts()
os_count = df['HĐH'].value_counts()
gender_count = df['Giới tính'].value_counts()
df
```

```python
plt.figure(figsize=(4, 3))
plt.pie(x=dev_count.values, labels=dev_count.index, autopct='%.2f%%')
plt.title("Phân bố Dòng điện thoại")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(4, 3))
plt.pie(x=os_count.values, labels=os_count.index, autopct='%.2f%%')
plt.title("Phân bố Hệ điều hành")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(4, 3))
plt.pie(x=gender_count.values, labels=gender_count.index, autopct='%.2f%%')
plt.title("Phân bố giới tính")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
ax = sns.countplot(data=df, x='Dòng thiết bị', hue='Giới tính')
ax.tick_params(axis='x', labelrotation=15)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(4, 3))
ax = sns.countplot(data=df, x='HĐH', hue='Giới tính')
ax.tick_params(axis='x', labelrotation=15)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 3))
sns.boxplot(x='Tuổi', y='Thời gian sử dụng ứng dụng (phút/ngày)', hue='Giới tính', data=df)
plt.title('Thời gian sử dụng ứng dụng theo Độ tuổi và Giới tính')
plt.show()
```

```python
plt.figure(figsize=(8, 3))
ax = sns.scatterplot(x='Thời gian sử dụng màn hình (giờ/ngày)', 
                     y='Mức độ hao pin (mAh/ngày)', 
                     hue='Dòng thiết bị', data=df)
plt.title('Thời gian sử dụng màn hình và Độ hao pin theo Dòng thiết bị')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='Dòng thiết bị', 
            y='Số lượng ứng dụng cài đặt', 
            hue='HĐH', data=df)
plt.title('Số lượng ứng dụng cài đặt theo Dòng thiết bị và HĐH')
plt.show()
```

```python
plt.figure(figsize=(8, 3))
ax = sns.histplot(data=df, 
                  x='Mức độ sử dụng dữ liệu (MB/ngày)', 
                  hue='Giới tính', bins=30, kde=True)
plt.title('Phân bố mức độ sử dụng dữ liệu theo giới tính')
plt.show()
```

```python
plt.figure(figsize=(8, 3))
sns.histplot(data=df, 
             x='Thời gian sử dụng màn hình (giờ/ngày)', 
             hue='Giới tính', bins=30, kde=True)
plt.title('Phân bố thời gian sử dụng màn hình theo giới tính')
plt.show()
```
